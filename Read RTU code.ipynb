```python
def readcode(code):
    
    if code[0:5] == '50GCQ':
        Unit_Series_Model = 'Packaged Rooftop Heat Pump'
    else:
        Unit_Series_Model = 'Unknown'
        warnings.warn("Unknown Unit/Series/Model")
        
    print("Unit/Series/Model:", code[0:5], "=" , Unit_Series_Model)
    
    if code[5] == 'M':
        Refrig_Systems_Options = 'Two Stage Cooling'
    else:
        Refrig_Systems_Options = 'Unknown'
        warnings.warn("Unknown Refrig. Systems Options")
    
    print("Refrig. Systems Options:", code[5], "=" , Refrig_Systems_Options)
    
    if code[6:8] == '04':
        Cooling_Tons = '3 tons'
    elif code[6:8] == '05':
        Cooling_Tons = '4 tons'
    elif code[6:8] == '06':
        Cooling_Tons = '5 tons'
    else:
        Cooling_Tons = 'Unknown'
        warnings.warn("Unknown Cooling Tons")
    
    print("Cooling Tons:", code[6:8], "=" , Cooling_Tons)
    
    if code[8] == "A":
        Sensor_Options = 'None'
    elif code[8] == "B":
        Sensor_Options = 'Return Air (RA) Smoke Detector'
    elif code[8] == "C":
        Sensor_Options = 'Supply Air (SA) Smoke Detector'
    elif code[8] == "D":
        Sensor_Options = 'RA + SA Smoke Detector'
    elif code[8] == "E":
        Sensor_Options = 'CO2 Sensor'
    elif code[8] == "F":
        Sensor_Options = 'RA Smoke Detector and CO2 Sensor'
    elif code[8] == "G":
        Sensor_Options = 'SA Smoke Detector and CO2 Sensor'
    elif code[8] == "H":
        Sensor_Options = 'RA + SA Smoke Detector and CO2 Sensor'
    elif code[8] == "J":
        Sensor_Options = 'Condensate Overflow Switch'
    elif code[8] == "K":
        Sensor_Options = 'Condensate Overflow Switch and RA Smoke Detector'
    elif code[8] == "L":
        Sensor_Options = 'Condensate Overflow Switch and RA and SA Smoke Detectors'
    elif code[8] == "M":
        Sensor_Options = 'Condensate Overflow Switch and SA Smoke Detector'
    else:
        Sensor_Options = 'Unknown'
        warnings.warn("Unknown Sensor Options")
    
    print("Sensor Options:", code[8], "=" , Sensor_Options)
    
    if code[9] == "1":
        Indoor_Fan_Options = 'Direct Drive — EcoBlue — Standard Static'
    elif code[9] == "2":
        Indoor_Fan_Options = 'Direct Drive — EcoBlue — Medium Static'
    elif code[9] == "3":
        Indoor_Fan_Options = 'Direct Drive — EcoBlue — High Static'
    else:
        Indoor_Fan_Options = 'Unknown'
        warnings.warn("Unknown Indoor Fan Options")

    print("Indoor Fan Options:", code[9], "=" , Indoor_Fan_Options)
    
    if code[10] == "A":
        Coil_Options = 'Al/Cu — Al/Cu'
    elif code[10] == "B":
        Coil_Options = 'Precoat Al/Cu — Al/Cu'
    elif code[10] == "C":
        Coil_Options = 'E-coat Al/Cu — Al/Cu'
    elif code[10] == "D":
        Coil_Options = 'E-coat Al/Cu — E-coat Al/Cu'
    elif code[10] == "E":
        Coil_Options = 'Cu/Cu — Al/Cu'
    elif code[10] == "F":
        Coil_Options = 'Cu/Cu — Cu/Cu'
    elif code[10] == "M":
        Coil_Options = 'Al/Cu — Al/Cu — Louvered Hail Guard'
    elif code[10] == "N":
        Coil_Options = 'Precoat Al/Cu — Al/Cu — Louvered Hail Guard'
    elif code[10] == "P":
        Coil_Options = 'E-coat Al/Cu — Al/Cu — Louvered Hail Guard'
    elif code[10] == "Q":
        Coil_Options = 'E-coat Al/Cu — E-coat Al/Cu — Louvered Hail Guard'
    elif code[10] == "R":
        Coil_Options = 'Cu/Cu — Al/Cu — Louvered Hail Guard'
    elif code[10] == "S":
        Coil_Options = 'Cu/Cu — Cu/Cu — Louvered Hail Guard'
    else:
        Coil_Options = 'Unknown'
        warnings.warn("Unknown Coil Options")

    print("Coil Options – (Outdoor - Indoor - Hail Guard):", code[10], "=" , Coil_Options)
    
    if code[11] == "1":
        Voltage = '575-3-60'
    elif code[11] == "3":
        Voltage = '208/230-1-60'
    elif code[11] == "5":
        Voltage = '208/230-3-60'
    elif code[11] == "6":
        Voltage = '460-3-60'
    else:
        Voltage = 'Unknown'
        warnings.warn("Unknown Voltage")
    
    print("Voltage:", code[11], "=" , Voltage)
    
    if code[12] == "-":
        Design_Revision = 'Factory Design Revision'
    else:
        Design_Revision = 'Unknown'
        warnings.warn("Unknown Design Revision")
    
    print("Design Revision:", code[12], "=" , Design_Revision)
    
    if code[13] == "0":
        Base_Unit_Controls = 'Electromechanical Controls — can be used with field-installed W7212 EconoMi$er® IV (Non-Fault Detection and Diagnostic)'
    elif code[13] == "2":
        Base_Unit_Controls = 'RTU Open Multi-Protocol Controller'
    elif code[13] == "3":
        Base_Unit_Controls = 'SystemVuTM Controls'
    elif code[13] == "6":
        Base_Unit_Controls = 'Electromechanical Controls — can be used with W7220 EconoMi$er X (with Fault Detection and Diagnostic)'
    else:
        Base_Unit_Controls = 'Unknown'
        warnings.warn("Unknown Base Unit Controls")
    
    print("Base Unit Controls:", code[13], "=" , Base_Unit_Controls) 
    
    if code[14] == "A":
        Intake_Exhaust_Options = 'None'
    elif code[14] == "B":
        Intake_Exhaust_Options = 'Temperature Economizer with Barometric Relief'
    elif code[14] == "F":
        Intake_Exhaust_Options = 'Enthalpy Economizer with Barometric Relief'
    elif code[14] == "K":
        Intake_Exhaust_Options = 'Two-Position Damper'
    elif code[14] == "U":
        Intake_Exhaust_Options = 'Temperature Ultra Low Leak Economizer with Barometric Relief'
    elif code[14] == "W":
        Intake_Exhaust_Options = 'Enthalpy Ultra Low Leak Economizer with Barometric Relief'
    else:
        Intake_Exhaust_Options = 'Unknown'
        warnings.warn("Unknown Intake / Exhaust Options")

    print("Intake / Exhaust Options:", code[14], "=" , Intake_Exhaust_Options)
    
    if code[15] == "0":
        Service_Options = 'None'
    elif code[15] == "1":
        Service_Options = 'Unpowered Convenience Outlet'
    elif code[15] == "2":
        Service_Options = 'Powered Convenience Outlet'
    elif code[15] == "3":
        Service_Options = 'Hinged Access Panels'
    elif code[15] == "4":
        Service_Options = 'Hinged Access Panels and Unpowered Convenience Outlet'
    elif code[15] == "5":
        Service_Options = 'Hinged Access Panels and Powered Convenience Outlet'
    elif code[15] == "6":
        Service_Options = 'MERV-8 High Efficiency Filters'
    elif code[15] == "7":
        Service_Options = 'Unpowered Convenience Outlet and MERV-8 High Efficiency Filters'
    elif code[15] == "8":
        Service_Options = 'Powered Convenience Outlet and MERV-8 High Efficiency Filters'
    elif code[15] == "9":
        Service_Options = 'Hinged Access Panels and MERV-8 High Efficiency Filters'
    elif code[15] == "A":
        Service_Options = 'Hinged Access Panels and Unpowered Convenience Outlet and MERV-8 High Efficiency Filters'
    elif code[15] == "B":
        Service_Options = 'Hinged Access Panels and Powered Convenience Outlet and MERV-8 High Efficiency Filters'
    elif code[15] == "C":
        Service_Options = 'Foil Faced Insulation'
    else:
        Service_Options = 'Unknown'
        warnings.warn("Unknown Service Options")
    
    print("Service Options:", code[15], "=" , Service_Options)
    
    if code[16] == "A":
        Electrical_Options = 'None'
    elif code[16] == "C":
        Electrical_Options = 'Non-Fused Disconnect (NFD)'
    elif code[16] == "D":
        Electrical_Options = 'Thru-The-Base Connections (TTB)'
    elif code[16] == "F":
        Electrical_Options = 'Non-Fused Disconnect and TTB Connections'
    elif code[16] == "N":
        Electrical_Options = 'Phase Monitor/Protection'
    elif code[16] == "Q":
        Electrical_Options = 'Phase Monitor/Protection and NFD'
    elif code[16] == "R":
        Electrical_Options = 'Phase Monitor/Protection and TTB Connections'
    elif code[16] == "T":
        Electrical_Options = 'Phase Monitor/Protection and NFD and TTB Connections'
    else:
        Electrical_Options = 'Unknown'
        warnings.warn("Unknown Electrical Options")
    
    print("Electrical Options:", code[16], "=" , Electrical_Options)
    
    if code[17] == "0":
        Packaging_Seismic_Compliance = 'Standard'
    elif code[17] == "1":
        Packaging_Seismic_Compliance = 'LTL'
    else:
        Packaging_Seismic_Compliance = 'Unknown'
        warnings.warn("Unknown Packaging & Seismic Compliance")
    
    print("Packaging & Seismic Compliance:", code[17], "=" , Packaging_Seismic_Compliance)
```
